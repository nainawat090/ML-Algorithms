```python
# KNN--->knn is a lasy instance based ml model  that is used for classification and regression the main goal is to predict the output
# for a one point find the k closest neighbours in the training data and use there neighbours. majority vote for
# clasification and average for regression.

# HOW KNN WORKS---> 1. store the training data then choose k nearest and a distance matrix then preprocess the features ,scale and standardize numeric feature
# then for each test sample compute distance from the test samples to every training sample. theen sort the training samples by distance
# and pic the top k. then for classification take the majority label among those k neighbours and for regression take the average of the neighbours.
# target values.
```

```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
irish_dataframe = pd.DataFrame(x_train,columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe , color by y_train
pd.plotting.scatter_matrix(irish_dataframe,c=y_train,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
prediction =knn.predict(x_test)
print("Prediction:",prediction)
print("Predicted target name:",iris_dataset['target_names'][prediction])
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df['fever']=si.fit_transform(df[['fever']])
```

```python
import numpy as np
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    df.drop(columns=["has_covid"]),df["has_covid"],test_size=.02,random_state=0)
```

```python
x_train
```

```python
pd.plotting.scatter_matrix(x_train,c=y_train,figsize=(15,15),
                          marker='o',hist_kwds={'bins':20},s=60,
                          alpha=.8,cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
