```python
 `/8import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv",
               usecols=['Age','EstimatedSalary','Purchased'])
```

```python
df.head(2)
```

```python
x=df.drop(columns=['Purchased'])
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_new=sc.fit_transform(x_train)
```

```python
x_test_new=sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier =GaussianNB()
```

```python
classifier.fit(x_train_new,y_train)
```

```python
y_pred=classifier.predict(x_test_new)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn=confusion_matrix(y_test,y_pred)
```

```python
cn
```

```python
# [[tp, fp]
#  [fn ,tn]]
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score
# Assuming y_true and y_tred are true and predicted labels respectively
# or we can say in bewteen total positive prediction how many are true

precision=precision_score(y_test,y_pred)
```

```python
precision
```

```python
from sklearn.metrics import recall_score
# Assuming y_true and y_pred are the true and predicted lables respectively
```

```python
recall=recall_score(y_test,y_pred)
```

```python
recall
```

```python
from sklearn.metrics import f1_score
# in total number of positive prediction how many are right
```

```python
f1 = f1_score(y_test,y_pred)
```

```python
f1
```
